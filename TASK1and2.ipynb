```python
trade=pd.read_csv("trades_file (1).csv")  #trade file
price=pd.read_csv("prices_file (1).csv")  #price file
```

```python
trade["tradevalue"]=trade["tradeSize"]*trade["tradePrice"]   #adding tradevalue column,as product of tradesize and tradeprice
trade
```

```python
a=pd.DataFrame(trade.groupby("symbol")["tradeSize"].sum())  #making dataframe on basis of individual symbol group and there tradesize sum in a day
a
```

```python
d=pd.DataFrame(trade.groupby("symbol")["tradevalue"].sum())  #making dataframe on basis of individual symbol group and there tradevalue sum in a day
d
```

```python
d.info()
```

```python
e=pd.merge(a,d,on="symbol",how="inner")   #making dataframe by merging above two data frame a,d on the basis of symbol and applying inner  join
e
```

```python
price  # price table
```

```python
b=pd.DataFrame(price[price['time']=="15:25:00"])  #filtering and making a dataframe where stock which are tarde after or on 15:25:00
b
```

```python
c = pd.merge(a, b, on=['symbol'], how='inner')   #merging dataframe a and b on basis of symbol so as to add time column in new data frame
c
```

```python
c["total_trade_value"]=c["tradeSize"]*c["price"]   #adding  column as total trade value by multiplying tradeszie and price at 15:25:00
c
```

```python
f=pd.merge(c,e,on="symbol",how="inner")   #creating a new dataframe and merging them on basis of symbol 
f
```

```python
f["PNL"]=f["tradevalue"]-f["total_trade_value"]   # adding final column of PNL and if PNL>0 it means profit and if PNL<0 it means loss
f
```

```python
f["L/P"]=f["tradevalue"]-f["total_trade_value"]   #adding column for showing profit and loss
for ind in f.index:
    if f['PNL'][ind]>0:
        f['L/P'][ind]="Profit"
    else:
        f['L/P'][ind]="Loss"
f
```
